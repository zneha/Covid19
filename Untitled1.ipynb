```python
df_needed = df[['Debt', 'Education', 'YearsEMployed', 'PriorDef', 'Employed', 'CreditScore', 'Income','Approved']]

cols_to_transform = [ 'Education', 'PriorDef', 'Employed', 'Approved']
df_needed_with_dummies = pd.get_dummies(df_needed, columns = cols_to_transform )
df_needed_with_dummies.head(100)
```

```python
# remove outliers
df_needed_with_dummies = df_needed_with_dummies.mask((df_needed_with_dummies - df_needed_with_dummies.mean()).abs() > 2 * df_needed_with_dummies.std()).dropna()
```

```python
from sklearn.model_selection import train_test_split
X = df_needed_with_dummies.loc[:, df_needed_with_dummies.columns != 'Approved_+']
y = df_needed_with_dummies[['Approved_+']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(X)
print(y)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 5, 
                             random_state = 0)

clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
